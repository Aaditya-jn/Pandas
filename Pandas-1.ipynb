Pandas Series -> A pandas series is like a column in a table. It is a 1D array holding data of any type.

Now we will learn to create Series

1) Series from lists ->

```python
import numpy as np
import pandas as pd

country = ["India", "Russia", "UAE", "USA", "Nepal"]

pd.Series(country)
```

From this we can see that, a series has the value and the index associated with it.
Also the data type of all the values is object because Pandas call strings as objects in 99% cases

```python
runs = [11, 13, 56, 65, 89]
pd.Series(runs)
```

We can also replace our indexes like in the below ->

```python
subjects = ["Maths", "Science", "Hindi", "English"]
marks = [78, 89, 88, 90]

pd.Series(marks, index = subjects)
```

We can also set up the name of our Series

```python
pd.Series(marks, index = subjects, name = "Aaditya's marks")
```

4) Sort values -> It sorts out the values of the dataset in ascending order

```python
kohli = (pd.read_csv("/home/aaditya/Downloads/kohli_ipl.csv", index_col = "match_no")).squeeze()

print(kohli.sort_values())
print(kohli.sort_values(ascending = False))
```

Edit items inside a series -> 

```python
import pandas as pd
mark_s = {
    "Maths" : 50,
    "Hindi" : 55,
    "Computer" : 44,
}

marks_series = pd.Series(mark_s, name = "Mere number")
marks_series
```

1] If we want to increase the marks of the subject which is at the 2nd index postion then ->

```python
marks_series[2] = 58
marks_series
```

2] If index postion does not exist then ->

```python
marks_series["EVS"] = 60
marks_series
```

3] We can do editing of series by slicing too ->

```python
marks_series[0:2] = [60, 60]
marks_series
```
