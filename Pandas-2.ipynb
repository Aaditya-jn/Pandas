11) Rename -> It is used to rename the particular column of the dataframe
This is a temporary change, for permanent change -> inplace = True

```python
import pandas as pd

student_data = [
    [100, 80, 10],
    [120, 90, 12],
    [80, 60, 7],
    [90, 70, 9]
]

df = pd.DataFrame(student_data, columns = ["IQ", "Marks", "Package"])

df = df.rename(columns={"Marks": "Percentage", "Package": "LPA"})

print(df)
```

```python
# Find all the final winners
win = ipl["MatchNumber"] == "Final"
mask = ipl[win]
mask[["Season", "WinningTeam"]]
```

```python
# How many super over finishes have occured

ipl[ipl["SuperOver"] == "Y"].shape
# So 14 that is no.of rows is the answer
```

```python
# How many matches has csk won in kolkata
ipl[(ipl["City"] == "Kolkata") & (ipl["WinningTeam"] == "Chennai Super Kings")].shape[0]
```

```python
# Toss winner is match winner in percentage

(ipl[ipl["TossWinner"] == ipl["WinningTeam"]].shape[0]/ipl.shape[0]) * 100

# 51 percent chances
```
